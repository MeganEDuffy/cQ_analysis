### Plotting WY22 and WY23 winter and non-winter event NO<sub>3</sub><sup>-</sup> and total phosphorus (TP) yields, NO<sub>3</sub><sup>-</sup>/TP to water yield ratios, and cumulative yields. 

### Yields were extracted in this notebook: https://github.com/MeganEDuffy/cQ_analysis/blob/main/millar2021_R_separation_hysteresis/sonya-may24/SV-2024.05.24-data_Millar_custom-input.ipynb

- This code delineates both purely automated events (using the RDF approach and based on Millar et al 2021 R workflow) and also takes as input manually (custom) delineated events, which works best for winter storms.

- The SV dataset is year-round, so in this notebook I use the RDF events for non-winter and the custom events for winter.
 - I manually changed the TP column in wb_q_chem.csv (from SV 2024-05-24) to be 'TP' (as for the Hungerford file) rather than 'TP_mgL' to be consistent. 


### Contextualizaing data: Kincaid et al., 2020 Hungerford and Wade data (2014-2018 non-winter events) with yields extracted here (similar to above) in this notebook: https://github.com/MeganEDuffy/cQ_analysis/blob/main/millar2021_R_separation_hysteresis/kincaid2020_hydroshare/Kincaid2020_Millar_custom-input.ipynb

- The Kincaid 2020 dataset has NO3 and SRP but not TP

```r
################
# LOAD PACKAGES #
################

library(tidyverse)
library(viridis)
library(lubridate)
library(patchwork)

###################
# SET DIRECTORIES #
###################

setwd("~//OneDrive/git-repos/EMMA")

input_dir <- "soil-resins/"
output_dir <- "~//OneDrive/git-repos/cQ_analysis/event-yields/"

# custom event delineation scripts puts original yields here:
SV_2021_2023_input_dir <- "~//OneDrive/git-repos/cQ_analysis/millar2021_R_separation_hysteresis/sonya-may24/custom-events_output/"
DK_2014_2018_yield_input_dir <- "~//OneDrive/git-repos/cQ_analysis/millar2021_R_separation_hysteresis/kincaid2020_hydroshare/output/"

################
# READ IN DATA #
################

# Read in cumulative 2021-2023 winter NO3 yield data
# Notebook here: https://github.com/MeganEDuffy/cQ_analysis/blob/main/millar2021_R_separation_hysteresis/sonya-may24/SV-2024.05.24-data_Millar_custom-input.ipynb
hungerford_yield_data <- read.csv(file.path(SV_2021_2023_input_dir, "Hungerford_NO3_2021-2023_StormEventSummaryDataCustomDelineations.csv")) %>%
  mutate(Date = as.Date(start)) %>% # Convert 'start' to Date format (drop time)
  rename(NO3_yield_mm = constit_yield_mm)

wade_yield_data <- read.csv(file.path(SV_2021_2023_input_dir, "Wade_NO3_2021-2023_StormEventSummaryDataCustomDelineations.csv")) %>%
  mutate(Date = as.Date(start)) %>% 
  rename(NO3_yield_mm = constit_yield_mm) 

# Read in cumulative 2021-2023 winter TP yield data
# Notebook here: https://github.com/MeganEDuffy/cQ_analysis/blob/main/millar2021_R_separation_hysteresis/sonya-may24/SV-2024.05.24-data_Millar_custom-input.ipynb
hungerford_yield_TP_data <- read.csv(file.path(SV_2021_2023_input_dir, "Hungerford_TP_2021-2023_StormEventSummaryDataCustomDelineations.csv")) %>%
  mutate(Date = as.Date(start)) %>% 
  rename(TP_yield_mm = constit_yield_mm)

wade_yield_TP_data <- read.csv(file.path(SV_2021_2023_input_dir, "Wade_TP_2021-2023_StormEventSummaryDataCustomDelineations.csv")) %>%
  mutate(Date = as.Date(start)) %>%
  rename(TP_yield_mm = constit_yield_mm)

# Read in Millar-delineated 2021-2023 non-ice NO3 yield and water yield data
# Same notebook as the custom input, but it also runs the Millar event delineation, then we'll separate out non-winter further down
# Just the 0.996 RDF results with SFT = 0.1
hungerford_yield_data_yearly_2123 <- read.csv(file.path(SV_2021_2023_input_dir, "Hungerford_NO3_2021-2023_StormEventSummaryData.csv")) %>%
  mutate(Date = as.Date(start)) %>% 
  rename(NO3_yield_mm = constit_yield_mm)

wade_yield_data_yearly_2123 <- read.csv(file.path(SV_2021_2023_input_dir, "Wade_NO3_2021-2023_StormEventSummaryData.csv")) %>%
  mutate(Date = as.Date(start)) %>% 
  rename(NO3_yield_mm = constit_yield_mm)

# Read in Millar-delineated 2021-2023 non-ice TP yield and water yield data
hungerford_yield_TP_data_yearly_2123 <- read.csv(file.path(SV_2021_2023_input_dir, "Hungerford_TP_2021-2023_StormEventSummaryData.csv")) %>%
  mutate(Date = as.Date(start)) %>%
  rename(TP_yield_mm = constit_yield_mm)

# Just the 0.996 RDF results with SFT = 0.05
wade_yield_TP_data_yearly_2123 <- read.csv(file.path(SV_2021_2023_input_dir, "Wade_TP_2021-2023_StormEventSummaryData.csv")) %>%
  mutate(Date = as.Date(start)) %>%
  rename(TP_yield_mm = constit_yield_mm)

# Read in Kincaid 2014-2018 non-ice NO3 yield data
# Notebook here: https://github.com/MeganEDuffy/cQ_analysis/blob/main/millar2021_R_separation_hysteresis/kincaid2020_hydroshare/Kincaid2020_Millar_custom-input.ipynb
hungerford_yield_data_yearly_1418 <- read.csv(file.path(DK_2014_2018_yield_input_dir, "Hungerford_NO3_2014-2018_StormEventSummaryDataCustomDelineations.csv")) %>%
  mutate(Date = as.Date(start))  %>%
  rename(NO3_yield_mm = constit_yield_mm)

wade_yield_data_yearly_1418 <- read.csv(file.path(DK_2014_2018_yield_input_dir, "Wade_NO3_2014-2018_StormEventSummaryDataCustomDelineations.csv")) %>%
  mutate(Date = as.Date(start)) %>%
  rename(NO3_yield_mm = constit_yield_mm)

####################
# MERGE NO3 and TP #
####################

# Select only 'storm_id' and 'TP_yield_mm' from TP datasets
hungerford_TP_subset <- hungerford_yield_TP_data %>%
  select(storm_id, TP_yield_mm)

wade_TP_subset <- wade_yield_TP_data %>%
  select(storm_id, TP_yield_mm)

hungerford_TP_subset_yearly_2123 <- hungerford_yield_TP_data_yearly_2123 %>%
  select(storm_id, TP_yield_mm)

wade_TP_subset_yearly_2123 <- wade_yield_TP_data_yearly_2123 %>%
  select(storm_id, TP_yield_mm)

# Merge TP data into the corresponding NO3 datasets using 'storm_id' as the key
hungerford_yield_data <- hungerford_yield_data %>%
  left_join(hungerford_TP_subset, by = "storm_id")

wade_yield_data <- wade_yield_data %>%
  left_join(wade_TP_subset, by = "storm_id")

hungerford_yield_data_yearly_2123 <- hungerford_yield_data_yearly_2123 %>%
  left_join(hungerford_TP_subset_yearly_2123, by = "storm_id")

wade_yield_data_yearly_2123 <- wade_yield_data_yearly_2123 %>%
  left_join(wade_TP_subset_yearly_2123, by = "storm_id")

###############
# FILTER DATA #
###############

# Don't need to filter Kincaid 2014-2018 dataset, that is all ice-free events

# Define Water Year 2022
wy22_start <- as.Date("2021-10-01")
wy22_end <- as.Date("2022-09-30")

# Define Water Year 2023
wy23_start <- as.Date("2022-10-01")
wy23_end <- as.Date("2023-09-30")

# Define Winter Water Year 2022
wwy22_start <- as.Date("2021-12-01")
wwy22_end <- as.Date("2022-05-01")

# Define Winter Water Year 2023
wwy23_start <- as.Date("2022-12-01")
wwy23_end <- as.Date("2023-05-01")

# Filter SV winter data for Winter Water Year 2022
hungerford_yield_data_wy22 <- hungerford_yield_data %>%
  filter(Date >= wwy22_start & Date <= wwy22_end) %>%
  mutate(Site = "Hungerford") %>%
  mutate(Season = "winter") %>%
  mutate(WY = '2022')

# Filter SV winter data for Winter Water Year 2023
hungerford_yield_data_wy23 <- hungerford_yield_data %>%
  filter(Date >= wwy23_start & Date <= wwy23_end) %>%
  mutate(Site = "Hungerford") %>%
  mutate(Season = "winter") %>%
  mutate(WY = '2023')

# Filter SV winter data for Winter Water Year 2022
wade_yield_data_wy22 <- wade_yield_data %>%
  filter(Date >= wwy22_start & Date <= wwy22_end) %>%
  mutate(Site = "Wade") %>%
  mutate(Season = "winter") %>%
  mutate(WY = '2022')

# Filter SV winter data for Winter Water Year 2023
wade_yield_data_wy23 <- wade_yield_data %>%
  filter(Date >= wwy23_start & Date <= wwy23_end) %>%
  mutate(Site = "Wade") %>%
  mutate(Season = "winter") %>%
  mutate(WY = '2023')

# Filter SV non-winter data for WY23
hungerford_yield_data_yearly_23 <- hungerford_yield_data_yearly_2123 %>%
  filter(Date >= wy23_start & Date <= wy23_end) %>%
  mutate(Site = "Hungerford") %>%
  mutate(Season = "non-winter") %>%
  mutate(WY = '2023')

# Filter SV non-winter data for WY23
wade_yield_data_yearly_23 <- wade_yield_data_yearly_2123 %>%
  filter(Date >= wy23_start & Date <= wy23_end) %>%
  mutate(Site = "Wade") %>%
  mutate(Season = "non-winter") %>%
  mutate(WY = '2023')

# Filter SV non-winter data for WY22
hungerford_yield_data_yearly_22 <- hungerford_yield_data_yearly_2123 %>%
  filter(Date >= wy22_start & Date <= wy22_end) %>%
  mutate(Site = "Hungerford") %>%
  mutate(Season = "non-winter") %>%
  mutate(WY = '2022')

# Filter SV non-winter data for WY22
wade_yield_data_yearly_22 <- wade_yield_data_yearly_2123 %>%
  filter(Date >= wy22_start & Date <= wy22_end) %>%
  mutate(Site = "Wade") %>%
  mutate(Season = "non-winter") %>%
  mutate(WY = '2022')

# Add Site, Season, and WY to Kincaid 2020 data
hungerford_yield_data_yearly_1418 <- hungerford_yield_data_yearly_1418 %>%
  mutate(Site = "Hungerford") %>%
  mutate(Season = "non-winter") %>%
  mutate(WY = '2014-2018')

# Add Site, Season, and WY to Kincaid 2020 data
wade_yield_data_yearly_1418 <- wade_yield_data_yearly_1418 %>%
  mutate(Site = "Wade") %>%
  mutate(Season = "non-winter") %>%
  mutate(WY = '2014-2018')

# Read in the compiled monthly available NH4/NO3 data (PO4 until 2023)
resin_data <- read.csv(file.path(input_dir, "Resin_timeseries.csv")) %>%
  mutate(Date = paste(Month, "1", Year, sep = " "),
         Date = mdy(Date))  # Convert the concatenated string to date format

# Filter data for Water Year 2023
resin_data_wy23 <- resin_data %>%
  filter(Date >= wy23_start & Date <= wy23_end)

# Filter data for Winter Water Year 2023
resin_data_winter_wy23 <- resin_data %>%
  filter(Date >= wwy23_start & Date <= wwy23_end)
```

## Plot WY22 WY23 winter and non-winter yields for NO3, TP
## Hungerford & Wade

```r
## Plot WY22 WY23 winter and non-winter yields for NO3 and TP
## Hungerford & Wade

# Combine the Hungerford datasets
combined_data_hungerford <- bind_rows(hungerford_yield_data_wy22, hungerford_yield_data_wy23, 
                                      hungerford_yield_data_yearly_22, hungerford_yield_data_yearly_23, 
                                      hungerford_yield_data_yearly_1418)

# Combine the Wade datasets
combined_data_wade <- bind_rows(wade_yield_data_wy22, wade_yield_data_wy23, 
                                wade_yield_data_yearly_22, wade_yield_data_yearly_23, 
                                wade_yield_data_yearly_1418)

# Fit linear models to calculate R² values
hungerford_winter22_lm <- lm(NO3_yield_mm ~ water_yield_mm, 
                    data = filter(combined_data_hungerford, Season == "winter", WY =='2022'))
hungerford_winter22_r2 <- summary(hungerford_winter22_lm)$r.squared

hungerford_winter23_lm <- lm(NO3_yield_mm ~ water_yield_mm, 
                    data = filter(combined_data_hungerford, Season == "winter", WY =='2023'))
hungerford_winter23_r2 <- summary(hungerford_winter23_lm)$r.squared

hungerford_nonwinter_lm <- lm(NO3_yield_mm ~ water_yield_mm, 
              data = filter(combined_data_hungerford, Season == "non-winter"))
hungerford_nonwinter_r2 <- summary(hungerford_nonwinter_lm)$r.squared

# Fit linear models to calculate R² values
wade_winter22_lm <- lm(NO3_yield_mm ~ water_yield_mm, 
                    data = filter(combined_data_wade, Season == "winter", WY =='2022'))
wade_winter22_r2 <- summary(wade_winter22_lm)$r.squared

wade_winter23_lm <- lm(NO3_yield_mm ~ water_yield_mm, 
                    data = filter(combined_data_wade, Season == "winter", WY =='2023'))
wade_winter23_r2 <- summary(wade_winter23_lm)$r.squared

wade_nonwinter_lm <- lm(NO3_yield_mm ~ water_yield_mm, 
              data = filter(combined_data_wade, Season == "non-winter"))
wade_nonwinter_r2 <- summary(wade_nonwinter_lm)$r.squared

# Combine the Hungerford datasets
combined_data_hungerford_nokincaid <- bind_rows(hungerford_yield_data_wy22, hungerford_yield_data_wy23)

# Combine the Wade datasets
combined_data_wade_nokincaid <- bind_rows(wade_yield_data_wy22, wade_yield_data_wy23)

# Combine the Hungerford datasets
combined_data_hungerford_nokincaid <- bind_rows(hungerford_yield_data_wy22, hungerford_yield_data_wy23, 
                                      hungerford_yield_data_yearly_22, hungerford_yield_data_yearly_23)

# Combine the Wade datasets
combined_data_wade_nokincaid <- bind_rows(wade_yield_data_wy22, wade_yield_data_wy23, 
                                wade_yield_data_yearly_22, wade_yield_data_yearly_23)

# Fit linear models to calculate R² values
hungerford_TPwinter22_lm <- lm(TP_yield_mm ~ water_yield_mm, 
                    data = filter(combined_data_hungerford_nokincaid, Season == "winter", WY =='2022'))
hungerford_TPwinter22_r2 <- summary(hungerford_TPwinter22_lm)$r.squared

hungerford_TPwinter23_lm <- lm(TP_yield_mm ~ water_yield_mm, 
                    data = filter(combined_data_hungerford_nokincaid, Season == "winter", WY =='2023'))
hungerford_TPwinter23_r2 <- summary(hungerford_TPwinter23_lm)$r.squared

hungerford_TPnonwinter_lm <- lm(TP_yield_mm ~ water_yield_mm, 
              data = filter(combined_data_hungerford_nokincaid, Season == "non-winter"))
hungerford_TPnonwinter_r2 <- summary(hungerford_TPnonwinter_lm)$r.squared

# Fit linear models to calculate R² values
wade_TPwinter22_lm <- lm(TP_yield_mm ~ water_yield_mm, 
                    data = filter(combined_data_wade_nokincaid, Season == "winter", WY =='2022'))
wade_TPwinter22_r2 <- summary(wade_TPwinter22_lm)$r.squared

wade_TPwinter23_lm <- lm(TP_yield_mm ~ water_yield_mm, 
                    data = filter(combined_data_wade_nokincaid, Season == "winter", WY =='2023'))
wade_TPwinter23_r2 <- summary(wade_TPwinter23_lm)$r.squared

wade_TPnonwinter_lm <- lm(TP_yield_mm ~ water_yield_mm, 
              data = filter(combined_data_wade_nokincaid, Season == "non-winter"))
wade_TPnonwinter_r2 <- summary(wade_TPnonwinter_lm)$r.squared

#############################
# Plot NO3 subplots A and C #
#############################

plot_A <- ggplot(combined_data_hungerford, aes(x = water_yield_mm, y = NO3_yield_mm)) +
  geom_point(
    aes(
      color = Season,
      shape = ifelse(Season == "winter" & WY == "2023", "Winter 2023", "Other")
    ),
    size = 7, alpha = 0.8
  ) +
  geom_smooth(
    data = filter(combined_data_hungerford, Season == "winter", WY == "2022"),
    method = lm, color = "#CD3333", linetype = "solid", se = TRUE
  ) +
  geom_smooth(
    data = filter(combined_data_hungerford, Season == "winter", WY == "2023"),
    method = lm, color = "#B22222", linetype = "solid", se = TRUE
  ) +
  geom_smooth(
    data = filter(combined_data_hungerford, Season == "non-winter"),
    method = lm, color = "black", linetype = "dotdash", se = TRUE
  ) +
  scale_color_manual(
    values = c(
      "winter" = "#CD3333",  # Winter
      "non-winter" = "black"
    ),
    guide = "none"  # Hides the color legend
  ) +
  scale_shape_manual(
    values = c(
      "Winter 2023" = 16,   # Square
      "Other" = 18          # Circle
    ),
    guide = "none"  # Hides the shape legend
  ) +
  labs(
    x = "Event water yield (mm)",
    y = "Event nitrate yield (kg NO3 km-2)",
    title = "A. Hungerford NO3",
    color = "Season"
  ) +
  xlim(0, 45) +
  ylim(0, 200) +
  theme_minimal(base_size = 20) +
  annotate(
    "text", x = 10, y = 10,
    label = paste0("Winter WY22 event R²: ", round(hungerford_winter22_r2, 3)),
    color = "#CD3333", hjust = 0, size = 8
  ) +
  annotate(
    "text", x = 2, y = 140,
    label = paste0("Winter WY23 event R²: ", round(hungerford_winter23_r2, 3)),
    color = "#B22222", hjust = 0, size = 8
  ) +
  annotate(
    "text", x = 0, y = 130,
    label = paste0("Non-winter event R²: ", round(hungerford_nonwinter_r2, 3)),
    color = "black", hjust = 0, size = 8
  )

plot_C <-  ggplot(combined_data_wade, aes(x = water_yield_mm, y = NO3_yield_mm)) +
  geom_point(
    aes(
      color = Season,
      shape = ifelse(Season == "winter" & WY == "2023", "Winter 2023", "Other")
    ),
    size = 7, alpha = 0.8
  ) +
  geom_smooth(
    data = filter(combined_data_wade, Season == "winter", WY == "2022"),
    method = lm, color = "darkolivegreen4", linetype = "solid", se = TRUE
  ) +
  geom_smooth(
    data = filter(combined_data_wade, Season == "winter", WY == "2023"),
    method = lm, color = "darkolivegreen4", linetype = "solid", se = TRUE
  ) +
  geom_smooth(
    data = filter(combined_data_wade, Season == "non-winter"),
    method = lm, color = "black", linetype = "dotdash", se = TRUE
  ) +
  scale_color_manual(
    values = c(
      "winter" = "darkolivegreen4",  # Winter
      "non-winter" = "black"
    ),
    guide = "none"  # Hides the color legend
  ) +
  scale_shape_manual(
    values = c(
      "Winter 2023" = 16,   # Square
      "Other" = 18          # Circle
    ),
    guide = "none"  # Hides the shape legend
  ) +
  labs(
    x = "Event water yield (mm)",
    y = "Event nitrate yield (kg NO3 km-2)",
    title = "C. Wade NO3",
    color = "Season"
  ) +
  #xlim(0, 250) +
  #ylim(0, 55) +
  theme_minimal(base_size = 20) +
  annotate(
    "text", x = 100, y = 10,
    label = paste0("Winter WY22 event R²: ", round(wade_winter22_r2, 3)),
    color = "darkolivegreen4", hjust = 0, size = 8
  ) +
  annotate(
    "text", x = 20, y = 35,
    label = paste0("Winter WY23 event R²: ", round(wade_winter23_r2, 3)),
    color = "darkolivegreen4", hjust = 0, size = 8
  ) +
  annotate(
    "text", x = 15, y = 40,
    label = paste0("Non-winter event R²: ", round(wade_nonwinter_r2, 3)),
    color = "black", hjust = 0, size = 8
  )

#############################
# Plot TP subplots B and D #
#############################

plot_B <-  ggplot(combined_data_hungerford_nokincaid, aes(x = water_yield_mm, y = TP_yield_mm)) +
  geom_point(
    aes(
      color = Season,
      shape = ifelse(Season == "winter" & WY == "2023", "Winter 2023", "Other")
    ),
    size = 7, alpha = 0.8
  ) +
  geom_smooth(
    data = filter(combined_data_hungerford_nokincaid, Season == "winter", WY == "2022"),
    method = lm, color = "#CD3333", linetype = "solid", se = TRUE
  ) +
  geom_smooth(
    data = filter(combined_data_hungerford_nokincaid, Season == "winter", WY == "2023"),
    method = lm, color = "#B22222", linetype = "solid", se = TRUE
  ) +
  geom_smooth(
    data = filter(combined_data_hungerford_nokincaid, Season == "non-winter"),
    method = lm, color = "black", linetype = "dotdash", se = TRUE
  ) +
  scale_color_manual(
    values = c(
      "winter" = "#CD3333",  # Winter
      "non-winter" = "black"
    )
  ) +
  scale_shape_manual(
    values = c(
      "Winter 2023" = 16,   # Square
      "Other" = 18          # Circle
    ),
    guide = "none"  # Hides the shape legend
  ) +
  labs(
    x = "Event water yield (mm)",
    y = "Event TP yield (kg TP km-2)",
    title = "B. Hungerford total P",
    color = "Season"
  ) +
  xlim(0, 45) +
  ylim(0, 20) +
  theme_minimal(base_size = 20) +
  annotate(
    "text", x = 10, y = 15,
    label = paste0("Winter WY22 event R²: ", round(hungerford_TPwinter22_r2, 3)),
    color = "#CD3333", hjust = 0, size = 8
  ) +
  annotate(
    "text", x = 2, y = 17,
    label = paste0("Winter WY23 event R²: ", round(hungerford_TPwinter23_r2, 3)),
    color = "#B22222", hjust = 0, size = 8
  ) +
  annotate(
    "text", x = 5, y = 12,
    label = paste0("Non-winter event R²: ", round(hungerford_TPnonwinter_r2, 3)),
    color = "black", hjust = 0, size = 8
  )

plot_D <-  ggplot(combined_data_wade_nokincaid, aes(x = water_yield_mm, y = TP_yield_mm)) +
  geom_point(
    aes(
      color = Season,
      shape = ifelse(Season == "winter" & WY == "2023", "Winter 2023", "Other")
    ),
    size = 7, alpha = 0.8
  ) +
  geom_smooth(
    data = filter(combined_data_wade_nokincaid, Season == "winter", WY == "2022"),
    method = lm, color = "darkolivegreen4", linetype = "solid", se = TRUE
  ) +
  geom_smooth(
    data = filter(combined_data_wade_nokincaid, Season == "winter", WY == "2023"),
    method = lm, color = "darkolivegreen4", linetype = "solid", se = TRUE
  ) +
  geom_smooth(
    data = filter(combined_data_wade_nokincaid, Season == "non-winter"),
    method = lm, color = "black", linetype = "dotdash", se = TRUE
  ) +
  scale_color_manual(
    values = c(
      "winter" = "darkolivegreen4",  # Winter
      "non-winter" = "black"
    )
  ) +
  scale_shape_manual(
    values = c(
      "Winter 2023" = 16,   # Circle
      "Other" = 18          # diamond
    ),
    guide = "none"  # Hides the shape legend
  ) +
  labs(
    x = "Event water yield (mm)",
    y = "Event TP yield (kg TP km-2)",
    title = "D. Wade total P",
    color = "Season"
  ) +
  #xlim(0, 250) +
  #ylim(0, 55) +
  theme_minimal(base_size = 20) +
  annotate(
    "text", x = 70, y = 0.9,
    label = paste0("Winter WY22 event R²: ", round(wade_TPwinter22_r2, 3)),
    color = "darkolivegreen4", hjust = 0, size = 8
  ) +
  annotate(
    "text", x = 75, y = 0.25,
    label = paste0("Winter WY23 event R²: ", round(wade_TPwinter23_r2, 3)),
    color = "darkolivegreen4", hjust = 0, size = 8
  ) +
  annotate(
    "text", x = 73, y = 0.1,
    label = paste0("Non-winter event R²: ", round(wade_TPnonwinter_r2, 3)),
    color = "black", hjust = 0, size = 8
  )

# Arrange the plots in a 2x2 grid
final_plot <- (plot_A | plot_B) / (plot_C | plot_D)

options(repr.plot.width = 20, repr.plot.height = 20) 

# Print the final plot
print(final_plot)

# Save combined plot
ggsave(plot = final_plot, width = 20, 
       height = 20, dpi = 300, 
       file = file.path(output_dir, "WY22_WY23_Hungerford_Wade_NO3_TP_yield_ratios.png"))
```

```r
library(dplyr)
library(purrr)

# Define a list of Hungerford datasets
hungerford_datasets <- list(hungerford_yield_data_wy22, hungerford_yield_data_wy23, 
                            hungerford_yield_data_yearly_22, hungerford_yield_data_yearly_23, 
                            hungerford_yield_data_yearly_1418)

# Define a list of Wade datasets
wade_datasets <- list(wade_yield_data_wy22, wade_yield_data_wy23, 
                      wade_yield_data_yearly_22, wade_yield_data_yearly_23, 
                      wade_yield_data_yearly_1418)

# Combine datasets using purrr::reduce()
combined_data_hungerford <- reduce(hungerford_datasets, bind_rows)
combined_data_wade <- reduce(wade_datasets, bind_rows)

# Define a function to fit linear models and extract R²
fit_r2 <- function(data, season, year = NULL) {
  filtered_data <- data %>% filter(Season == season, if (!is.null(year)) WY == year else TRUE)
  model <- lm(NO3_yield_mm ~ water_yield_mm, data = filtered_data)
  summary(model)$r.squared
}

# Apply function to Hungerford dataset
hungerford_r2 <- list(
  winter22 = fit_r2(combined_data_hungerford, "winter", "2022"),
  winter23 = fit_r2(combined_data_hungerford, "winter", "2023"),
  nonwinter = fit_r2(combined_data_hungerford, "non-winter")
)

# Apply function to Wade dataset
wade_r2 <- list(
  winter22 = fit_r2(combined_data_wade, "winter", "2022"),
  winter23 = fit_r2(combined_data_wade, "winter", "2023"),
  nonwinter = fit_r2(combined_data_wade, "non-winter")
)

# Print results
hungerford_r2
wade_r2
```

```r
library(ggplot2)
library(patchwork)

# Define a function to generate the plots dynamically
create_plot <- function(data, title, y_var, r2_values, colors) {
  ggplot(data, aes(x = water_yield_mm, y = .data[[y_var]])) +
    geom_point(
      aes(
        color = Season,
        shape = ifelse(Season == "winter" & WY == "2023", "Winter 2023", "Other")
      ),
      size = 7, alpha = 0.8
    ) +
    geom_smooth(
      data = filter(data, Season == "winter", WY == "2022"),
      method = lm, color = colors["winter22"], linetype = "solid", se = TRUE
    ) +
    geom_smooth(
      data = filter(data, Season == "winter", WY == "2023"),
      method = lm, color = colors["winter23"], linetype = "solid", se = TRUE
    ) +
    geom_smooth(
      data = filter(data, Season == "non-winter"),
      method = lm, color = colors["nonwinter"], linetype = "dotdash", se = TRUE
    ) +
    scale_color_manual(values = colors, guide = "none") +
    scale_shape_manual(values = c("Winter 2023" = 16, "Other" = 18), guide = "none") +
    labs(
      x = "Event water yield (mm)",
      y = ifelse(y_var == "NO3_yield_mm", "Event nitrate yield (kg NO3 km-2)", "Event TP yield (kg TP km-2)"),
      title = title,
      color = "Season"
    ) +
    theme_minimal(base_size = 20) +
    annotate("text", x = 10, y = 10, label = paste0("Winter WY22 event R²: ", round(r2_values$winter22, 3)), color = colors["winter22"], hjust = 0, size = 8) +
    annotate("text", x = 25, y = 40, label = paste0("Winter WY23 event R²: ", round(r2_values$winter23, 3)), color = colors["winter23"], hjust = 0, size = 8) +
    annotate("text", x = 0, y = 130, label = paste0("Non-winter event R²: ", round(r2_values$nonwinter, 3)), color = colors["nonwinter"], hjust = 0, size = 8)
}

# Define color schemes
hungerford_colors <- c("winter22" = "#CD3333", "winter23" = "#B22222", "nonwinter" = "black")
wade_colors <- c("winter22" = "darkolivegreen4", "winter23" = "darkolivegreen4", "nonwinter" = "black")

# Create plots using the function
plot_A <- create_plot(combined_data_hungerford, "A. Hungerford NO3", "NO3_yield_mm", hungerford_r2, hungerford_colors)
plot_B <- create_plot(combined_data_hungerford, "B. Hungerford total P", "TP_yield_mm", hungerford_r2, hungerford_colors)
plot_C <- create_plot(combined_data_wade, "C. Wade NO3", "NO3_yield_mm", wade_r2, wade_colors)
plot_D <- create_plot(combined_data_wade, "D. Wade total P", "TP_yield_mm", wade_r2, wade_colors)

# Arrange the plots in a 2x2 grid
final_plot <- (plot_A | plot_B) / (plot_C | plot_D)

options(repr.plot.width = 20, repr.plot.height = 20) 

# Print and save the final plot
print(final_plot)

ggsave(plot = final_plot, width = 20, height = 20, dpi = 300, 
       file = file.path(output_dir, "WY22_WY23_Hungerford_Wade_NO3_TP_yield_ratios.png"))
```
