```python
#####################
# READ IN LIBRARIES #
#####################

import pandas as pd  
import numpy as np  
import geopandas as gpd  
import matplotlib.pyplot as plt # plot our data and manipulate the plot
import contextily as ctx # add a default basemap under our polygons of interest
from shapely.geometry import Point, Polygon

############################
# VIEW CONTEXTILY BASEMAPS #
############################

# Using contextily for basemaps (https://dlab.berkeley.edu/news/adding-basemaps-python-contextily)
# change basemap - can be one of these
# first see available provider names

print(ctx.providers.keys())
print(ctx.providers.USGS.keys())

#####################################
# READ IN SHP FILES AND SITE COORDS #
#####################################

# Read in the shapefile of watershed polygons
wade = gpd.read_file('WadeShape/layers/WadeBrook.shp')
hungerford = gpd.read_file('HungerfordShape/layers/HungerfordShape.shp')

# Transform to ESRI:102039
wade_102039 = wade.to_crs("ESRI:102039")
hungerford_102039 = hungerford.to_crs("ESRI:102039")

# Load the site coordinates
site_coords = pd.read_csv("newernet-site-coodinates.csv")

# Filter for Hungerford and Wade watersheds
filtered_sites = site_coords[site_coords['Watershed'].isin(['Hungerford', 'Wade'])]

# Convert to a GeoDataFrame using X and Y coordinates
gdf_sites = gpd.GeoDataFrame(
    filtered_sites,
    geometry=gpd.points_from_xy(filtered_sites['X-coordinate'], filtered_sites['Y-coordinate']),
    crs="EPSG:32145"  # EPSG code for NAD_1983_StatePlane_Vermont_FIPS_4400
)

# Reproject sites to ESRI:102039
gdf_sites_102039 = gdf_sites.to_crs("ESRI:102039")

###########
# MAPPING #
###########

# Create the plot
fig, ax = plt.subplots(figsize=(10, 10))

# Plot the watersheds
wade_102039.plot(ax=ax, facecolor="none", edgecolor="blue", linewidth=3, label="Wade Watershed")
hungerford_102039.plot(ax=ax, facecolor="none", edgecolor="green", linewidth=3, label="Hungerford Watershed")

# Add basemap in the correct projection
ctx.add_basemap(ax, source=ctx.providers.USGS.USTopo, crs=wade_102039.crs)

# Add legend
ax.legend()
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx

# Load the site coordinates
site_coords = pd.read_csv("newernet-site-coodinates.csv")

# Filter for Hungerford and Wade watersheds
filtered_sites = site_coords[site_coords['Watershed'].isin(['Hungerford', 'Wade'])]

# Convert to a GeoDataFrame using X and Y coordinates
gdf_sites = gpd.GeoDataFrame(
    filtered_sites,
    geometry=gpd.points_from_xy(filtered_sites['X-coordinate'], filtered_sites['Y-coordinate']),
    crs="EPSG:32145"  # EPSG code for NAD_1983_StatePlane_Vermont_FIPS_4400
)

# Reproject sites to ESRI:102039
gdf_sites_102039 = gdf_sites.to_crs("ESRI:102039")

# Plotting the watersheds and sites together
fig, ax = plt.subplots(figsize=(12, 12))

# Plot the watershed polygons
wade_102039.plot(ax=ax, facecolor="none", edgecolor="blue", linewidth=3, label="Wade Watershed")
hungerford_102039.plot(ax=ax, facecolor="none", edgecolor="green", linewidth=3, label="Hungerford Watershed")

# Plot the site locations
gdf_sites_102039.plot(ax=ax, color='red', markersize=50, label="Sampling Sites")

# Add a basemap
ctx.add_basemap(ax, source=ctx.providers.USGS.USTopo, crs=gdf_sites_102039.crs)

# Add legend and title
ax.legend()
plt.title("Watersheds and Sampling Sites")
plt.show()
```

```python
# Unique watersheds to plot separately
watersheds_to_plot = ["Hungerford", "Wade"]

# Create a dictionary to map watershed names to their respective GeoDataFrames
watershed_gdfs = {
    "Hungerford": hungerford_102039,
    "Wade": wade_102039
}

# Set zoom axis coordinates for Hungerford
#plt.ylim([2651000,2654500])
#plt.xlim([1791000,1795000])

# Set zoom axis coordinates for Wade
#plt.ylim([2656000,2658000])
#plt.xlim([1830000,1832500])


# Filter sites by watershed
for watershed_name in watersheds_to_plot:
    # Filter the sampling sites for the current watershed
    watershed_sites = gdf_sites_102039[gdf_sites_102039['Watershed'] == watershed_name]
    
    # Access the corresponding watershed GeoDataFrame
    watershed_polygon = watershed_gdfs[watershed_name]
    
    # Create a new plot for the current watershed
    fig, ax = plt.subplots(figsize=(10, 10))
    
    # Plot the watershed polygon
    watershed_polygon.plot(ax=ax, facecolor="none", edgecolor="blue", linewidth=3, label=f"{watershed_name} watershed")
    
    # Plot the sites within the watershed
    watershed_sites.plot(ax=ax, color='black', markersize=50, marker='x', label="Sampling sites")
    
    # Add basemap
    ctx.add_basemap(ax, source=ctx.providers.USGS.USTopo, crs=watershed_polygon.crs)
    
    # Add title and legend
    ax.set_title(f"{watershed_name} watershed and sampling sites", fontsize=16)
    ax.legend()
    
    # Show the plot
    plt.show()
```
